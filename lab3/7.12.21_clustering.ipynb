```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

airline_df = pd.read_csv('https://raw.githubusercontent.com/wangrenfeng0/Machine-Learning-I/main/airline_passenger_satisfaction.csv') 

if 'Unnamed: 0' in airline_df:
    del airline_df['Unnamed: 0']
    
airline_df = airline_df.dropna()


Satisfaction ={'neutral or dissatisfied':0,'satisfied':1}
airline_df.satisfaction = [Satisfaction[item] for item in airline_df.satisfaction]

    
    
airline_df = pd.get_dummies(airline_df, columns=['Gender', 'customer_type','type_of_travel','customer_class'])
```

```python
airline_2 = airline_df.drop('satisfaction', axis=1).copy()
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
cv = StratifiedKFold(n_splits=10,shuffle=True, random_state=5)                     
print(cv)
```

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd


cls = KMeans(n_clusters=2, init='random',random_state=1)
cls.fit(airline_2)
newfeature = cls.labels_ # the labels from kmeans clustering
```

```python
airline_df['Cluster'] = cls.labels_
aa = airline_df[['Cluster','satisfaction']].copy()
aa['records_count'] = 1
aa.groupby(['Cluster','satisfaction']).count()
```

```python
from sklearn import metrics as mt
confMat = mt.confusion_matrix(airline_df['satisfaction'].values,airline_df['Cluster'].values)
acc = mt.accuracy_score(airline_df['satisfaction'],airline_df['Cluster'])

tn, fp, fn, tp = confMat.ravel()
    
specificity = tn / (tn+fp)
sensitivity = tp / (tp+fn)


print("confusion matrix\n",confMat)
print("accuracy", acc )
print('specificity', specificity)
print('sensitivity', sensitivity, '\n')
    
```
